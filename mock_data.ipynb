### Customer data: name, gender, age, destination category (forest, beach, city, â€¦), Temperature (cold, warm, hot), budget (low, medium, high), purpose (business, leisure), number of people wish to travel with, families (0 or 1), Duration, Travel months, transportation (train, airplane, car, bus, ferry)

```python
# Generate 10000 fake customer names using library
fake = Faker()
customer_names = [fake.name() for _ in range(10000)]
```

```python
# Define other columns categories
genders = ['Male', 'Female']
ages = list(range(18, 65))
destination_cats = ["Mountain","City",
    "Countryside","Island","Ski Resort","Village",
    "Lake","Winery","Spa Retreat","Photography Expedition"]
temperatures = ['Cold', 'Warm', 'Hot']
budgets = ['Low', 'Medium', 'High']
purposes = ['Business', 'Leisure']
num_people = list(range(1, 6))
families = [0, 1] # indicator; 1=Y, 0=N
durations = list(range(1, 15))
travel_months = ['September','October', 'November']
transportation = ['Train', 'Airplane', 'Car', 'Bus', 'Ferry']
language = ['EN', 'FR', 'ZH']
hobby = ['Creative Arts', 'Physical Sports', 'Outdoor Exploration', 'Culinary', 'Intellectual Pursuits']

# Identify number of customer records to generate
num_records = 10000

# Generate data and put into dictionary form
customer_data = []
for _ in range(num_records):
    record = {
        'Name': random.choice(customer_names),
        'Gender': random.choice(genders),
        'Age': random.choice(ages),
        'Destination_Category': random.choice(destination_cats),
        'Temperature': random.choice(temperatures),
        'Budget': random.choice(budgets),
        'Purpose': random.choice(purposes),
        'Num_of_People': random.choice(num_people),
        'Families': random.choice(families),
        'Duration': random.choice(durations),
        'Travel Months': random.choice(travel_months),
        'Transportation': random.choice(transportation),
        'Hobby': random.choice(hobby),
        "Language": random.choice(language)
    }
    customer_data.append(record)
```

```python
# Create a DataFrame
customer_df = pd.DataFrame(customer_data)

# Display the generated DataFrame
customer_df
```

```python
customer_df.to_csv('customer.csv')
```

### YUL to other cities airfare, round-way, one way per record, for 30days

```python
d2 = {'Flight':["YUL-YYZ",
               "YHZ-YYZ",
               "YVR-YYZ",
               "YOW-YYZ",
               "YWG-YYZ",
               "YXE-YYZ",
               "YYJ-YYZ",
               "YYC-YYZ",
               "YEG-YYZ",
               "YXY-YYZ",
               "YQR-YYZ",
               "YZF-YYZ"]
               #"Richmond","Hamilton","Sherbrooke","Niagara","Vaughan","Banff", "London"],
}
```

```python
d2.update(price)
```
